### 2.1 Cargo el corpus de 20 News Groups

```python
!pip install -q sentence-transformers


import os

path = "/kaggle/input/20-newsgroups"

documents = []

for filename in os.listdir(path):
    if filename.endswith(".txt"):
        with open(os.path.join(path, filename), "r", encoding="latin-1") as f:
            text = f.read()
            docs = text.split("\n\n")

            for d in docs:
                documents.append({
                    "file": filename,
                    "text": d
                })

len(documents)
```

```python
sample_docs = documents[:10000]
len(sample_docs)
```

```python
texts = [d["text"] for d in sample_docs]
```

### 2.2 Transformo a embeddings

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

embeddings = model.encode(texts)

print("Shape de embeddings:", embeddings.shape)
print(embeddings[0][:100])
```

### 2.3 Creo una query y hago la búsqueda

```python
import random

def mostrar_resumen_corpus(texts, k=5):
    print("\n== RESUMEN DEL CORPUS =\n")
    
    for i in random.sample(range(len(texts)), min(k, len(texts))):
        print(f"[ID: {i}]")
        print(texts[i][:30000])  
        print("\n----------------------------------------")

mostrar_resumen_corpus(texts, k=5)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

query = "Olympus Stylus"

query_emb = model.encode([query])

sims = cosine_similarity(query_emb, embeddings)[0]

top_k = 5
indices = np.argsort(sims)[::-1][:top_k]

print("\n==RESULTADOS DE LA BUSQUEDA ==\n")

for i in indices:
    doc = sample_docs[i]

    print(f" INDEX INTERNO: {i}")
    print(f" ARCHIVO: {doc['file']}")
    print(f" SCORE: {sims[i]:.4f}")
    print("\n TEXTO:")
    print(doc["text"][:60000])
    print("\n---------------------------\n")
```

Obtengo los 5 documentos más similares a mi query

```python
query = "Olympus Stylus".lower()

matches = []

for i, doc in enumerate(sample_docs):
    text = doc["text"].lower()

    if query in text:
        matches.append(i)

print("Total coincidencias exactas:", len(matches))

for idx in matches[:10]:   # mostramos solo los primeros 10
    print("\n===== MATCH =====")
    print(f"DOCUMENTO INDEX: {idx}")
    print(f"ARCHIVO: {sample_docs[idx]['file']}")
    print(sample_docs[idx]["text"][:400])
    print("\n-------------------")
```
