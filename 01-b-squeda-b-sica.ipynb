```python
#Parte 1: Búsqueda lineal en documentos

palabra = input("Escribe una palabra a buscar: ").lower()

inicio = time.time()
resultados = []
for i in range(len(df)):
    texto = str(df.loc[i, 'review']).lower()
    if palabra in texto.split():
        resultados.append(i)
fin = time.time()

print(f"La palabra '{palabra}' aparece en {len(resultados)} reseñas.")
print("Tiempo de búsqueda lineal:", round(fin - inicio, 4), "segundos")
```

```python
#Parte 2: Construcción de un índice invertido
indice = {}
for i in range(len(df)):
    palabras = str(df.loc[i, 'review']).lower().split()
    for p in palabras:
        if p not in indice:
            indice[p] = [i]
        elif i not in indice[p]:
            indice[p].append(i)

print("Índice invertido creado con éxito!")
```

```python
#Parte 3: Evaluación de tiempos de búsqueda
palabra = input("Palabra para comparar tiempos: ").lower()

t1 = time.time()
res_lineal = [i for i in range(len(df)) if palabra in str(df.loc[i, 'review']).lower().split()]
t2 = time.time()

t3 = time.time()
res_indice = indice.get(palabra, [])
t4 = time.time()

# Mostrar resultados
print(f"\nBúsqueda lineal encontró {len(res_lineal)} reseñas en {round(t2 - t1, 4)} segundos.")
print(f"Búsqueda con índice encontró {len(res_indice)} reseñas en {round(t4 - t3, 4)} segundos.")
```

```python
## Conclusión: Se comprobo que la busqueda lineal es más eficiente cuando se tiene un índice invertido y los tiempo de respuesta es más corto.
## Ojo: para la comprobación de tiempos entre los dos datasets se descomento las lineas del comienzo del código y para validar tiempo se rejudo la la lista para minimizar tiempos de carga.
```
