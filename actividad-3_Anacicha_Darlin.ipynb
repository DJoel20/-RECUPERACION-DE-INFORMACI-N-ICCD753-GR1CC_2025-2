## 1. Limpiar los documentos de caracteres que no corresponden


```python
doc = df.iloc[0]['review']
doc
```

```python
def limpiar_texto(doc):
    doc = re.sub(r"<.*?>", " ", doc)      # eliminar HTML
    doc = re.sub(r"[^a-zA-Z\s]", " ", doc) # eliminar símbolos
    doc = re.sub(r"\s+", " ", doc)         # eliminar espacios extra
    return doc.lower().strip()

df['clean'] = df['review'].apply(limpiar_texto)

df['clean'].iloc[0]
```

```python
df['clean'].iloc[0].split()
```

```python
def normalize(text):
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)
    return text

df['normalized'] = df['clean'].apply(normalize)
df['normalized'].iloc[0]
```

## Normalización 

Convertir todos los tokens a minúsculas.

Elimina puntuación y símbolos no alfabéticos.

```python
def normalize(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

df['normalized'] = df['clean'].apply(normalize)
```

## Eliminación de Stopwords

Eliminar las palabras vacías (stopwords) usando una lista estándar de la librería nitk. 

```python
import nltk
nltk.data.path.append('/kaggle/input/stopwordws/stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))
print("Stopwords cargadas:", len(stop_words))
```

```python
def remove_stopwords(text):
    tokens = word_tokenize(text)
    return [w for w in tokens if w.lower() not in stop_words]

df['no_stopwords'] = df['normalized'].apply(remove_stopwords)

df['no_stopwords'].iloc[0][:20]
```

## Stemming
Reducir el espacio de palabras


```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

def apply_stemming(tokens):
    return [stemmer.stem(w) for w in tokens]

df['stemmed'] = df['no_stopwords'].apply(apply_stemming)

df['stemmed'].iloc[0][:20]

```

## Verificar

Comparar el tamaño del diccionario de términos del corpus antes y después de aplicar el preprocesamiento



```python
vocab_original = set(" ".join(df['review']).lower().split())

vocab_final = set(w for tokens in df['stemmed'] for w in tokens)

print("Tamaño vocabulario original :", len(vocab_original))
print("Tamaño vocabulario final    :", len(vocab_final))
```
